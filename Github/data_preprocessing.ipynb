```python
import re
def find_non_capitalized(text):
    word_list = text.split(" ")
    for word in word_list:
        if not word[0].isupper():
            return word

# Apply the function to the 'desc' column
data['first_non_capitalized'] = data['desc'].apply(find_non_capitalized)
```

```python
def find_blocking_player(row):
    if row['first_non_capitalized'] == 'blocks':
        word_list = row['desc'].split(" ")
        player_name = ""
        for word in word_list:
            if word[0].isupper():
                player_name = player_name + word + " "
            else:
                break
        player_name = player_name[:-1]
        return player_name
data['blocking_player'] = data.apply(lambda row: find_blocking_player(row),axis =1)
```

```python
def find_shooting_player(row):
    word_list = row['desc'].split(" ")
    if row['first_non_capitalized'] != 'blocks':
        player_name = ""
        for word in word_list:
            if word[0].isupper():
                player_name = player_name + word + " "
            else:
                break
        player_name = player_name[:-1]
    else:
        player_name = ""
        for word in word_list[word_list.index('blocks')+1:]:
            if word[0].isupper():
                player_name = player_name + word + " "
            else:
                break
        player_name = player_name[:-1].replace("'s","")
    return player_name
```

# **Shot value**

```python
data.columns
```

The shot value is always 0 points in case of a miss. It could prove useful to quickly know when a shot would have been worth 1,2 or 3 points in case of having been made. There are bound to be some mistaken imputations due to the integer format of the coordinates and shot characteristics (if the feet of the shooter leave the ground outside the line...)

```python
import numpy as np

def shot_value(x, y):
    """
    Determine the shot value based on coordinates (x, y).
    
    Parameters:
    x (int): Horizontal position (0 to 50)
    y (int): Vertical position (0 to 47)
    
    Returns:
    int: 2 or 3 depending on shot zone
    """
    # Hoop coordinates
    hoop_x, hoop_y = 25, 5
    
    # Calculate distance from hoop
    distance = np.sqrt((x - hoop_x) ** 2 + (y - hoop_y) ** 2)
    
    # WNBA 3-point range thresholds
    three_point_distance_center = 22.15 # 3-point arc in the center
    three_point_distance_sides = 22.0    # Shorter arc on the sides
    
    # Check if within 3-point range
    if y > 14 and distance > three_point_distance_center:
        return 3
    elif y <= 14 and distance > three_point_distance_sides:
        return 3
    else:
        return 2

```

```python
def impute_max_points(row):
    if row['grouped_shot_type'] == 'Free Throw':
        return 1
    if row['shot_value']!=0:
        return row['shot_value']
    return shot_value(row['coordinate_x'], row['coordinate_y'])
```

```python
data['max_value'] = data.apply(impute_max_points,axis=1)
```
